## Creates simple render of a DRAG pulse 

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x= np.linspace(0, 40e-9, 500)
amp = 1
mu=max(x/2)
f_pulse_mod=100e6
t=x 
sigma = 5e-9 
motzoi = .5

gauss_env = amp*np.exp(-(0.5 * ((t-mu)**2) / sigma**2))
deriv_gauss_env = motzoi * -1 * (t-mu)/(sigma**1) * gauss_env

gauss_env_I_mod = gauss_env* np.cos(f_pulse_mod*2*np.pi*t)
gauss_env_Q_mod = gauss_env* np.sin(f_pulse_mod*2*np.pi*t)

deriv_gauss_env_I_mod =deriv_gauss_env* np.cos(f_pulse_mod*2*np.pi*t)
deriv_gauss_env_Q_mod =deriv_gauss_env* np.sin(f_pulse_mod*2*np.pi*t)
```

```python
f, ax=plt.subplots()
ax.plot(t,deriv_gauss_env, c='C1', ls='--', lw=2, label='deriv')
ax.plot(t,-deriv_gauss_env, c='C1', ls='--', lw=2)
ax.plot(t,deriv_gauss_env_I_mod, c='C1', lw=2)
ax.plot(t,deriv_gauss_env_Q_mod, c='C1', lw=2)

ax.plot(t, gauss_env, c='C0', ls='--', lw=2, label='Gauss')
ax.plot(t, -gauss_env, c='C0', ls='--', lw=2)
ax.plot(t, gauss_env_I_mod, c='C0', lw=2)
ax.plot(t, gauss_env_Q_mod, c='C0', lw=2)

plt.axis('off')
f.savefig('DRAG_pulse.svg', transparent=True, dpi=600)
# plt.legend()
```
